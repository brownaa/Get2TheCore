## Convert empty trays to np.NaN

```python
for row in range(len(train.OutputID)):
    for r in range(1,12):
        x = 0
        y = 0
        for i in range(1,5):
            x += train.at[row,f'row_{r}_X_{i}']
            y += train.at[row,f'row_{r}_Y_{i}']
        if x + y == 0:
            for i in range(1,5):
                train.loc[[row],[f'row_{r}_X_{i}']] = np.NaN
                train.loc[[row],[f'row_{r}_Y_{i}']] = np.NaN
```

```python
#train2 = pd.read_csv('train_with_labels_v3.csv')
for r in range(1,12):
    merged_rows = []
    for row in range(len(train.OutputID)):
        merged_row = []
        for i in range(1,5):
            merged_row.append(train.at[row,f'row_{r}_X_{i}'])
            merged_row.append(train.at[row,f'row_{r}_Y_{i}'])
        merged_rows.append(merged_row)
    train[f'row_{r}'] = merged_rows
    train.drop(list(train.filter(regex = f'row_{r}_')), axis = 1, inplace = True) 
```

```python
def mk_bbox(height, width, coords, bbox_frac = 0.03):
    min_dim = min(height, width)
    offset = int(round(min_dim * bbox_frac))
    
    for i in range(int(len(coords)/2)):
        x = coords[::2]
        y = coords[1::2]
        d = {"Corner Upper Left":[max(0, x[0]-offset),         #X upper left
                           max(0, height-y[0]-offset),  #Y upper left
                           min(width, x[0]+offset),           #X lower right
                           min(height, height-y[0]+offset)],   #Y lower right
             "Corner Upper Right":[max(0, x[1]-offset),         #X upper left
                           max(0, height-y[1]-offset),  #Y upper left
                           min(width, x[1]+offset),           #X lower right
                           min(height, height-y[1]+offset)],   #Y lower right
             "Corner Lower Right":[max(0, x[2]-offset),         #X upper left
                           max(0, height-y[2]-offset),  #Y upper left
                           min(width, x[2]+offset),           #X lower right
                           min(height, height-y[2]+offset)],   #Y lower right
             "Corner Lower Left":[max(0, x[3]-offset),         #X upper left
                           max(0, height-y[3]-offset),  #Y upper left
                           min(width, x[3]+offset),           #X lower right
                           min(height, height-y[3]+offset)]   #Y lower right
            }
    return(d)
```
