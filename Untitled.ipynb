## Convert empty trays to np.NaN

```python
for row in range(len(train.OutputID)):
    for r in range(1,12):
        x = 0
        y = 0
        for i in range(1,5):
            x += train.at[row,f'row_{r}_X_{i}']
            y += train.at[row,f'row_{r}_Y_{i}']
        if x + y == 0:
            for i in range(1,5):
                train.loc[[row],[f'row_{r}_X_{i}']] = np.NaN
                train.loc[[row],[f'row_{r}_Y_{i}']] = np.NaN
```

```python
train.head()
```

```python
#train2 = pd.read_csv('train_with_labels_v3.csv')
for r in range(1,12):
    merged_rows = []
    for row in range(len(train.OutputID)):
        merged_row = []
        for i in range(1,5):
            merged_row.append(train.at[row,f'row_{r}_X_{i}'])
            merged_row.append(train.at[row,f'row_{r}_Y_{i}'])
        merged_rows.append(merged_row)
    train[f'row_{r}'] = merged_rows
    train.drop(list(train.filter(regex = f'row_{r}_')), axis = 1, inplace = True) 
```
